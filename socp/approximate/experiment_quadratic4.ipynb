```julia
include("functions.jl")

function f(x)
    theta = [3,-5,0]
    c = exp(theta[1]+theta[2]*(x-theta[3])^2)
    f = [1,(x-theta[3])^2,2*theta[2]*(theta[3]-x) ]
    return sqrt(c)/(1+c)*f
end

ex1 = example(3, [-1,1], f)
```

```julia
misocp(200,3,ex1)
solve!(prob, MosekSolver())
```

```julia
n = [16; 32; 64; 128; 256; 512; 1024]
save = zeros(length(n), 4*2+2)
for i in 1:length(n)
    t1 = time()
    misocp(n[i],4,ex1)
    solve!(prob, MosekSolver())
    t2 = time()
    cc = w.value[w.value.>0]
    cc = [cc; x[getindex.(findall(x->x>0, w.value),1)]]
    cc = [cc; (prob.optval^4/0.02023431778515896)^(1/3)]
    cc = [cc; t2-t1]
    save[i,:] = cc
end
```
