```julia
function f(x)
    theta = [-2.163, -0.1]
    c = 1-exp(-30*exp(theta[1]+theta[2]*x))
    f = [1,x]
    return sqrt(c)*f
end
ex1 = example(2,[0,1],f)

n = [16; 32; 64; 128; 256; 512; 1024]
save = zeros(length(n), 2*2+2)
for i in 1:length(n)
    t1 = time()
    misocp(n[i],2,ex1)
    solve!(prob, MosekSolver())
    t2 = time()
    cc = w.value[w.value.>0]
    cc = [cc; x[getindex.(findall(x->x>0, w.value),1)]]
    cc = [cc; (prob.optval^2/0.23138226556240576)^(1/2)]
    cc = [cc; t2-t1]
    save[i,:] = cc
end
```
