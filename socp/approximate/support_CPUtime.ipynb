```julia
include("functions.jl")

function f(x)
    theta = [0.05884, 4.298, 21.8]
    c1 = x*theta[3]*exp(-theta[1]*x)
    c2 = -x*theta[3]*exp(-theta[2]*x) 
    c3 = exp(-theta[2]*x)-exp(-theta[1]*x)
    return [c1; c2; c3]
end

ex1 = example(3,[0,30],f)
```

```julia
misocp(128,3,ex1)
solve!(prob, MosekSolver(LOG=0))
```

```julia
using Statistics
n = 100
save = zeros(100)
for i in 3:100
    temp = zeros(100)
    for j in 1:100
        t1 = time()
        misocp(128,i,ex1)
        solve!(prob, MosekSolver(LOG=0))
        t2 = time()
        temp[j] = t2-t1
    end
    save[i] = median(temp)
    print(i)
end
```

```julia
using JLD2, FileIO
@save "exact_save.jld2" save
```
