```julia
include("functions.jl")

function f(x)
    theta = [0.05884, 4.298, 21.8]
    c1 = x*theta[3]*exp(-theta[1]*x)
    c2 = -x*theta[3]*exp(-theta[2]*x) 
    c3 = exp(-theta[2]*x)-exp(-theta[1]*x)
    return [c1; c2; c3]
end

ex1 = example(3,[0,30],f)
```

```julia
misocp(30,3,ex1)
solve!(prob, MosekSolver())
```

```julia
save = zeros(5)

for k = 1:5
    t1 = time()
    misocp(100,k,ex1)
    solve!(prob, MosekSolver())
    t2 = time()
    save[k] = t2-t1
end
```
