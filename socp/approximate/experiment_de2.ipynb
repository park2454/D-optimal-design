```julia
include("functions.jl")

function f(x)
    theta = [.8; .5; .5]
    c = theta[1]*exp(theta[2]*x) + (1-theta[1])*exp(-theta[3]*x)
    f = [1; (exp(theta[2]*x) - exp(-theta[3]*x))/c; theta[1]*x*exp(theta[2]*x)/c; -(1-theta[1])*x*exp(-theta[3]*x)/c]
    return f
end

ex1 = example(4,[0,10],f)
```

```julia
misocp(30,4,ex1)
solve!(prob, MosekSolver())
```

```julia
n = [16; 32; 64; 128; 256; 512; 1024]
save = zeros(length(n), 4*2+2)
for i in 1:length(n)
    t1 = time()
    misocp(n[i],4,ex1)
    solve!(prob, MosekSolver())
    t2 = time()
    cc = w.value[w.value.>0]
    cc = [cc; x[getindex.(findall(x->x>0, w.value),1)]]
    cc = [cc; (prob.optval^4/0.0009723832732216562)^(1/4)]
    cc = [cc; t2-t1]
    save[i,:] = cc
end
```
