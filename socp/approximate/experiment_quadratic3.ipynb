```julia
include("functions.jl")

function f(x)
    theta = [2,3,0]
    c = exp(theta[1]+theta[2]*(x-theta[3])^2)
    f = [1,(x-theta[3])^2,2*theta[2]*(theta[3]-x) ]
    return sqrt(c)/(1+c)*f
end

ex1 = example(3, [-3,1], f)
```

```julia
misocp(100,3,ex1)
solve!(prob, MosekSolver())
```

```julia
n = [16; 32; 64; 128; 256; 512; 1024]
save = zeros(length(n), 3*2+2)
for i in 1:length(n)
    t1 = time()
    misocp(n[i],3,ex1)
    solve!(prob, MosekSolver())
    t2 = time()
    cc = w.value[w.value.>0]
    cc = [cc; x[getindex.(findall(x->x>0, w.value),1)]]
    cc = [cc; (prob.optval^4/5.69464649495885e-5)^(1/3)]
    cc = [cc; t2-t1]
    save[i,:] = cc
end
```
