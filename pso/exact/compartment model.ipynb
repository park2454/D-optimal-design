```julia
function f(x)
    theta = [0.05884, 4.298, 21.8]
    c1 = x[1]*theta[3]*exp(-theta[1]*x[1])
    c2 = -x[1]*theta[3]*exp(-theta[2]*x[1]) 
    c3 = exp(-theta[2]*x[1])-exp(-theta[1]*x[1])
    return [c1; c2; c3]
end

N = 9
dim = 1
design = [0 30]
niter = 2^10
nptc = 2^10

t1 = time()
result = psox(N, f, dim, design, niter, nptc)
t2 = time()
t2-t1
```

```julia
N = 3*(1:20)
result = zeros(20)
t = zeros(20)
for i = 1:20
    t1 = time()
    result[i] = psox(N[i], f, dim, design, niter, nptc)[1]
    t2 = time()
    t[i] = t2 -t1
end

```
